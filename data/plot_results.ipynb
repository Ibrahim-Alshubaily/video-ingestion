```python
import os
import random
import matplotlib.pyplot as plt
import cv2

# Path to annotated image folder
annotated_dir = os.path.join("2025_05_28", "annotated")

image_files = [f for f in os.listdir(annotated_dir) if f.endswith(('.jpg', '.png'))]
selected = random.sample(image_files, min(6, len(image_files)))

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, img_file in enumerate(selected):
    img_path = os.path.join(annotated_dir, img_file)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axes[i].imshow(img)
    axes[i].set_title(img_file, fontsize=8)
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```
